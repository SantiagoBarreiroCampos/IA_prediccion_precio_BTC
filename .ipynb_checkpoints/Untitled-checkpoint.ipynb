# Aprendizaje automático

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
def split_label(df, test_size, label):
    train, test = train_test_split(df, test_size=test_size)
    features = df.columns.drop(label)
    
    train_X = train[features]
    train_y = train[label]
    test_X = test[features]
    test_y = test[label]
    
    return train_X, train_y, test_X, test_y   
```

```python
train_X, train_y, test_X, test_y = split_label(tweetsDF, 0.2, '1d_despues')

print(train_X.shape)
print(test_X.shape)

print(train_y.shape)
print(test_y.shape)
```

# Redes neuronales

```python
## Esta primera si tenemos datos categóricos.
from sklearn.neural_network import MLPClassifier
## Esta segunda si tenemos datos numéricos.
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
#tweetsDF = tweetsDF.sample(frac = 1).head(1000)
X = tweetsDF[['user_followers', 'sentimiento', '7d_atras', '1d_atras']]
Y = tweetsDF['7d_despues']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size =.2, random_state = 1)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(150,100,50), max_iter=300,activation = 'relu',solver='adam',random_state=1)

model = clf.fit(X_train, y_train)

predictions = model.predict(X_test)
print("R cuadrado: ", r2_score(y_test, predictions))

# Revisar los parámetros que podemos incluir en clf y experimentar con ellos. Ya que no hay una regla que diga lo que es mejor
# o peor en todas las ocasiones.
```

# K-Fold

```python
from sklearn.model_selection import KFold
from sklearn import linear_model
from sklearn.metrics import r2_score
import numpy as np
```

```python
tweetsDF
```

```python
X = tweetsDF[['1d_atras', '7d_atras']]
Y = tweetsDF['1d_despues']
```

```python
kf = KFold(n_splits = 10, shuffle = True)
kf.get_n_splits(X)
```

```python
regr = linear_model.LinearRegression()

# Creamos un vector donde vamos a estar guardando todos los resultados.
resultados = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X.loc[train_index,], X.loc[test_index,]
    y_train, y_test = Y[train_index], Y[test_index]
    regr.fit(X_train, y_train)
    predicciones = regr.predict(X_test)
    print("R2: ", r2_score(y_test, predicciones))
    resultados.append(r2_score(y_test, predicciones))
print("R2 medio: ", np.mean(resultados))
```
